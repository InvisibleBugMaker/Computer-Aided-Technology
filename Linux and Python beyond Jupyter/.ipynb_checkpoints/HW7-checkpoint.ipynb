7. Make a single matrix to rotate a 2D point around the point 5,5 by 45 degrees counter clockwise. Apply it to the point 10,10

```python
theta = np.radians(45)
cos = np.cos(theta)
sin = np.sin(theta)

pt = np.array([5,5])
m = np.array([10,10,1])
x = pt[0]
y = pt[1]
a = x-cos*x+sin*y
b = y- sin*x-cos*y
rm = np.array([[cos, -sin,a], [sin, cos,b],[0,0,1]])
result = np.dot(rm,m)[:2]
print('Rotation Matrix',rm)

print('Result is', result)
```

8. Make a 3D rotation by 20 degrees ,30 degrees ,40 degrees in the following coordinate systems
<ol type='a'>
<li>Z,X’, Z’’<\li>
<li>Y,X’Z’’<\li>
<li>Y,X’,Y’’<\li>
<li>Z, X, Y<\li>
<\ol>
    
* Apply them all to the points [0,0,1] , [0,1,0] and [1,0,0]

```python
def Rot_x(theta):
    '''get the rotation matrix of x'''
    cos = np.cos(theta)
    sin = np.sin(theta)
    rot_x = np.array([[1,0,0],[0,cos,-sin],[0,sin,cos]])
    return rot_x

def Rot_y(theta):
    '''get the rotation matrix of y'''
    cos = np.cos(theta)
    sin = np.sin(theta)
    rot_y = np.array([[cos,0,sin],[0,1,0],[-sin,0,cos]])
    return rot_y

def Rot_z(theta):
    '''get the rotation matrix of z'''
    cos = np.cos(theta)
    sin = np.sin(theta)
    rot_z = np.array([[cos,-sin,0],[sin,cos,0],[0,0,1]])
    return rot_z

point1 = np.array([0,0,1])
point2 = np.array([0,1,0])
point3 = np.array([1,0,0])
```

```python
#### A ####
rotate = Rot_z(np.radians(40))@Rot_x(np.radians(30))@Rot_z(np.radians(20))
print('[0,0,1] is', rotate@point1)
print('[0,1,0] is', rotate@point2)
print('[1,0,0] is', rotate@point3)
```

```python
#### B ####
rotate = Rot_y(np.radians(40))@Rot_x(np.radians(30))@Rot_z(np.radians(20))
print('[0,0,1] is', rotate@point1)
print('[0,1,0] is', rotate@point2)
print('[1,0,0] is', rotate@point3)
```

```python
#### C ####
rotate = Rot_y(np.radians(40))@Rot_x(np.radians(30))@Rot_y(np.radians(20))
print('[0,0,1] is', rotate@point1)
print('[0,1,0] is', rotate@point2)
print('[1,0,0] is', rotate@point3)
```

```python
#### D ####
rotate = Rot_z(np.radians(20))@Rot_x(np.radians(30))@Rot_y(np.radians(40))
print('[0,0,1] is', rotate@point1)
print('[0,1,0] is', rotate@point2)
print('[1,0,0] is', rotate@point3)
```

9. Using the plot functions I showed in class and axis angle rotations, rotate a cube 45 degrees along one of its corners. What is the axis you used? 

```python
#got it from the lecture note
%matplotlib notebook

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from itertools import product, combinations


def SO3plot(transform = np.eye(3)):
    '''get it from lecture notes'''
    fig = plt.figure()
    ax = fig.gca(projection='3d')

    lim = 3
    
    ax.set_xlim(-lim,lim)
    ax.set_ylim(-lim,lim)
    ax.set_zlim(-lim,lim)


        # draw axis
    
    for i in range(0,3):
        s = np.array([0,0,0])
        e = np.array([0,0,0])
        s[i]=lim
        e[i]=-lim
        ax.plot3D(*zip(s,e), color="k")


    
    # draw cube
    r = [-2, 2]
    for s, e in combinations(np.array(list(product(r, r, r))), 2):
        if np.sum(np.abs(s-e)) == r[1]-r[0]:
            ax.plot3D(*zip(transform@s, transform@e), color="r")


    N=50
    stride=1
    r = 1.0
    u = np.linspace(0, 2 * np.pi, N)
    v = np.linspace(0, np.pi, N)
    x = np.outer(r*np.cos(u), r*np.sin(v))
    y = np.outer(r*np.sin(u), r*np.sin(v))
    z = np.outer(np.ones(np.size(u)), r*np.cos(v))
    print(x.shape)
    ax.plot_surface(x, y, z, linewidth=0.0, cstride=stride, rstride=stride)
    
def Nx(n):
    '''get it from lecture notes'''
    x = n[0]
    y = n[1]
    z = n[2]
    nx = np.array( [ [0, -z,  y],
                     [z,  0,  x],
                     [-y,  x, 0] ] )
    return nx

def AxisAngle(n,theta):
    '''get it from lecture notes'''
    mag = np.linalg.norm(n)
    v = n/mag
    s = np.sin(theta)
    c = np.cos(theta)
    NM = Nx(v)
    r = np.eye(3)+s*NM+(1-c)*NM@NM
    return r

T = AxisAngle([1,1,-1],np.pi/4)
SO3plot(T)
```

10. Make a homogenous transformation matrix to rotate around the point 5,5,5 by 20 degrees along the[ 1,1,1] axis. Apply it to the points,( 10,10,10 ) (10,0,0), (0,10,0) and (0,0,10)

```python
#got this code from lecture note
def Nx(n):
    '''get it from lecture notes'''
    x = n[0]
    y = n[1]
    z = n[2]
    nx = np.array( [ [0, -z,  y],
                     [z,  0,  x],
                     [-y,  x, 0] ] )
    return nx

def AxisAngle(n,theta):
    '''get it from lecture notes'''
    mag = np.linalg.norm(n)
    v = n/mag
    s = np.sin(theta)
    c = np.cos(theta)
    NM = Nx(v)
    r = np.eye(3)+s*NM+(1-c)*NM@NM
    return r

def AxisAngleDistance(n,theta,D):
    '''get it from lecture notes'''
    R = AxisAngle(n,theta)
    d = np.array(D).T.reshape((1,3)) 
    return np.block([
                    [R , d.T],
                    [0,0,0, 1]
                    ])
n = [1,1,1]
theta = np.pi/9
D = [5,5,5]
p1 = AxisAngleDistance(n,theta,D)@[10,10,10,1]
print('[10,10,10] is',p1[0:3])
p2 = AxisAngleDistance(n,theta,D)@[10,0,0,1]
print('[10,0,0] is',p2[0:3])
p3 = AxisAngleDistance(n,theta,D)@[0,10,0,1]
print('[0,10,0] is',p3[0:3])
p4 = AxisAngleDistance(n,theta,D)@[0,0,10,1]
print('[0,0,10] is',p4[0:3])
```

12. Write a function that makes flail, it should take in three parameters M,N and R. It should have a spherical core of radius R,  and regularly space M pointy bits in N in rings around the sphere.

```python
def flail(M,N,R):
    '''Generate one line of bits first'''
    d = 0
    f = 0
    for i in range(1,N+1):
        c = rotate(180/(N+1)*i,[1,0,0])(cylinder(r1 = R/2,r2=0,h=R+2))
        d += c
    '''Then make the line around the sphere'''
    for j in range(M):
        e = rotate(360/M*j,[0,0,1])(d)
        f += e
    
    s = sphere(r = R)
    return r.render(d+s+f)

flail(5,3,5)
```

13 (2 pts) Make a program that takes in a 2D array of floats, and a tuple of voxel dimensions (x,y,z) and produces a height map with each cuboid being x by y and the float value tall. . It should ignore values less than one z voxel height and omit them from the object

```python
import numpy as np

def h_map(a,b):
    result = 0
    '''Get the position'''
    for i in range(len(a)):
        for j in range(len(a[0])):
            if a[i,j] < b[2]:
                '''check the height'''
                a[i,j] = b[2]
            c = translate([i*b[0],j*b[1]])(cube([b[0],b[1],a[i,j]]))
            result += c
    return r.render(result)

a = np.array([2,5,7.4,4,6,3.2,7.1,9.8,6]).reshape(3,3)
b = [3,2,2]
h_map(a,b)
```
