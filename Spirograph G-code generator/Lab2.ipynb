```python
import csv
import numpy as np
import math
import matplotlib.pyplot as plt

def make_list_1():
    r1 = 0.5
    r2 = 40.0/96.0
    r3 = 40.0/96.0

    ncycle = 5
    x1=[]
    y1=[]
    i = 0
    while i <= ncycle*2*np.pi:
        x1.append(np.real(r3*np.exp(1j*i*(r1+r2)/r2) + (r1+r2)*np.exp(1j*i))*40+125)
        y1.append(np.imag(r3*np.exp(1j*i*(r1+r2)/r2) + (r1+r2)*np.exp(1j*i))*40+125)
        i += 0.05
    return x1, y1
    
def make_gcode_1(a,b):
    with(open('spirograph111.gcode', 'w')) as f:
        me = csv.writer(f)
        me.writerow(["M140 S60"])
        me.writerow(["M105"])
        me.writerow(["M190 S60"])
        me.writerow(["M104 S230"])
        me.writerow(["M105"])
        me.writerow(["M109 S230"])
        me.writerow(["M82"])
        me.writerow(["G92 E0"])
        me.writerow(["G28"])
        me.writerow(["G91"])
        me.writerow(["G1 Z2.0 F3000"])
        me.writerow(["G1 X5 Y20 Z0.3 F5000.0"])
        me.writerow(["G92 E0"])
        me.writerow(["G92 E0"])
        me.writerow(["G1 F2700 E-5"])
        me.writerow(["M107"])
        
        
        
        for i in range(len(a)):
            me.writerow(["G1 " + "X" + str(a[i]) + " Y" + str(b[i])  + " Z0"])
        me.writerow(["G0 Z10"])
        me.writerow(["G28"])


def plot_spirograph(x_values,y_values):
    plt.plot(x_values,y_values)
    plt.axis('equal')
    plt.show()

    
x1,y1 = make_list_1()
make_gcode_1(x1,y1)
plot_spirograph(x1,y1)
```

```python
import csv
import numpy as np
import math
import matplotlib.pyplot as plt

def make_list_2():
    r1 = 0.5
    r2 = 40.0/96.0
    r3 = 40.0/96.0

    ncycle = 5
    t = np.linspace(0, ncycle*2*np.pi, 1000)
    x2 = np.real(r3*np.exp(1j*t*(r1+r2)/r2) + (r1+r2)*np.exp(1j*t))*40+125
    y2 = np.imag(r3*np.exp(1j*t*(r1+r2)/r2) + (r1+r2)*np.exp(1j*t))*40+125
    return x2, y2
    

def make_gcode_2(a,b):
    with(open('spirograph222.gcode', 'w')) as f:
        me = csv.writer(f)
        me.writerow(["G28"])
        me.writerow(["G0 Z10"])
        me.writerow(["G0 X50 Y50"])
        
        for i in range(len(a)):
            me.writerow(["G1 " + "X" + str(a[i]) + " Y" + str(b[i]) + " Z0"])
        me.writerow(["G0 Z10"])
        me.writerow(["G28"])
    
def plot_spirograph(x_values,y_values):
    plt.plot(x_values,y_values)
    plt.axis('equal')
    plt.show()

x2, y2 = make_list_2()
make_gcode_2(x2,y2)
plot_spirograph(x2,y2)
```
